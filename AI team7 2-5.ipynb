```python
# import
from google.colab import drive
import pandas as pd
import csv

# 구글 드라이브 mount
drive.mount('/content/gdrive')

# 파일 경로 설정
file_path = '/content/gdrive/MyDrive/Colab Notebooks/data-01-test-score.csv'
```

```python
class ReadCSV() :
  # 생성자
  def __init__(self, file_path) :
    self.file_path = file_path

  # 파일 읽는 함수
  def read_file(self, file_path) :
    data = []
    f = open(self.file_path, 'r')
    read = csv.reader(f)
    for line in read:
        data.append(line)
    f.close()

    return data

  # 리스트 값들의 평균 구하는 함수
  def merge_list(self, file_path) :

    # read_file 수행 시 불러와지는 data
    data = self.read_file(file_path)

    # 리스트 합 값을 저장하는 리스트 생성
    merge_data = []

    # 리스트 평균 값을 저장하는 리스트 생성
    avg_data = []

    sum = 0
    
    for subject_data in data :
      for personal_data in subject_data :
        # 리스트 각 값을 더하여 sum 값 생성
        sum += int(personal_data)

      # sum값을 merge_data에 append
      merge_data.append(sum)
      
      # 새로운 sum을 구하기 위해 초기화
      sum=0
    

    # 평균값 구하기
    for i in range(0,len(merge_data)) :
      avg_data.append(merge_data[i]/4)
    
    # 오름차순 정렬
    avg_data.sort()

    return avg_data
```

```python
read_csv = ReadCSV(file_path)
print(read_csv.merge_list(file_path))
```
