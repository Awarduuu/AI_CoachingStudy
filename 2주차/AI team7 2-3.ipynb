# Q3. 이번 시험 결과에 대한 데이터를 학과 사무실에서 CSV파일로 전달해줬습니다. 우리는 이 파일을 이용해서 데이터 처리를 진행해야 합니다. 파일 입출력을 이용해 파일 데이터를 리스트로 만들어 보세요.

```python
# 패키지 import
from google.colab import drive
import pandas as pd
import csv
```

```python
# 구글 드라이브 mount
drive.mount('/content/gdrive')
```

```python
# file_path 지정
# 경로 : 로컬 구글드라이브
file_path = '/content/gdrive/MyDrive/Colab Notebooks/data-01-test-score.csv'

def read_file(file_path):
    # 저장할 데이터 리스트 생성
    data = []

    # open함수를 이용해 csv 파일 읽기
    f = open(file_path, 'r')
    read = csv.reader(f)
    
    for line in read:
        data.append(line)
    f.close()

    return data
```

```python
print(read_file(file_path))
```
