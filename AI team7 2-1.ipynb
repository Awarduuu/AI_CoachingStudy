# Q1. 중간고사, 기말고사 점수를 따로 받아 저장하는 클래스를 구현해보세요. 단, 생성자의 인스턴스는 private으로 선언되어야 하며, 데코레이터를 이용해 데이터를 저장하고, 평균값을 출력해보세요.

```python
# test score, mid : 50, final : 70

class Score() :
  def __init__(self, mid, final):
    # 생성자 인스턴스 private으로 선언
    self.__mid = mid
    self.__final = final
  
  # Decorator
  @property 
  def mid(self):
    return self.__mid

  @property
  def final(self):
    return self.__final
```

```python
# 점수 저장
score = Score(50,75)

# 평균값 출력
print((score.mid + score.final) / 2)
```
