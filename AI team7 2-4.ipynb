```python
# import
from google.colab import drive
import pandas as pd
import csv

# 구글드라이브 mount
drive.mount('/content/gdrive')

# file_path 설정
file_path = '/content/gdrive/MyDrive/Colab Notebooks/data-01-test-score.csv'
```

```python
# ReadCSV 클래스 생성

class ReadCSV() :
  # 생성자
  def __init__(self, file_path) :
    self.file_path = file_path

  
  # 파일 읽는 함수 
  def read_file(self, file_path) :
    data = []
    f = open(self.file_path, 'r')
    read = csv.reader(f)
    for line in read:
        data.append(line)
    f.close()

    return data

  # 리스트 내 데이터 합 구하는 함수
  def merge_list(self, file_path) :
    data = self.read_file(file_path)
    merge_data = []

    sum = 0
    
    for subject_data in data :
      for personal_data in subject_data :
        sum += int(personal_data)
      merge_data.append(sum)
      sum=0
    
    return merge_data
```

```python
read_csv = ReadCSV(file_path)
```

```python
print(read_csv.read_file(file_path))
```

```python
print(read_csv.merge_list(file_path))
```
